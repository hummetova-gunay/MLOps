```python
# horizontally stacking
first= np.full((2,3),3)
print(first)
second = np.full((2,4),5)
print(second)
np.hstack([first, second])
```

```python
# vertical stacking
third = np.full((3,3), 1)
forth = np.full((3,3), 2)
np.vstack((third, forth))
```

```python
# splitting arrays
np.hsplit(second, 2)
```

```python
# unpacking
first, second = np.hsplit(second,2)
print(first)
print(second)
```
