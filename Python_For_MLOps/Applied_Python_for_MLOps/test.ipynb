# ML studio with a Service Principal

```python
import os
subscription_id = os.getenv('SUBSCRIPTION_ID')
resource_group = os.getenv('RESOURCE_GROUP', default='demo-try-azureml')
workspace_name = os.getenv('WORKSPACE_NAME', default= 'demo-try-azureml')
workspace_region = os.getenv('WORKSPACE_REGION', default='eastus2')
```

```python
from azureml.core import Workspace

try:
    ws = Workspace(subscription_id = subscription_id, resource_group = resource_group, workspace_name = workspace_name)
    ws.write_config()
    print('Workspace configuration succeed')
except Exception:
    print('Workspace not accessible. Ensure that Codespaces secrets are available, see the first cell')
    raise
```
